**Problem 1**: 10 pts - For the function

$$
f(x) = \frac{x^{2}}{1+x^{4}}, ~ -1 \leq x \leq 1
$$

a) (2pts) Using an equispaced set of $10$ nodes, generate the Lagrange interpolating polynomial to $f(x)$.  Generate a plot which shows how the error in your approximation varies over the interval.  

b) (2pts) Using an equispaced set of $20$ nodes, generate the Lagrange interpolating polynomial to $f(x)$.  Generate a plot which shows how the error in your approximation varies over the interval.  

c) (2pts) At what number of equispaced nodes does your Lagrange interpolation approximation break down?  

d) (4pts) Does using Chebyshev points help resolve the issues you saw in c) ?  Provide examples to verify your claim.

```python
# Function definition and definition of "true" information 
f = lambda x: x**2./(1.+x**4.)
xinterp = np.linspace(-1.,1.,int(1e3)+1)
ftrue = f(xinterp)
```

```python
# Lagrange Interpolation Implementation
def lfun(xvals,jj,x):
    lval = np.ones(x.size)
    xj = xvals[jj]
    xkk = np.concatenate((xvals[:jj],xvals[jj+1:]))
    for xk in xkk:
        lval *= (x-xk)/(xj-xk)
    return lval

def lagran_interp(xvals,fvals,x):
    ipoly = np.zeros(x.size)
    for jj in xrange(fvals.size):
        ipoly += fvals[jj]*lfun(xvals,jj,x)
    return ipoly

# Error for equi-spaced interpolation
def equi_spaced_error(skip,xinterp,ftrue):
    nmpts = np.int(np.round(int(1e3)/skip))
    xvals = xinterp[::nmpts]
    fvals = ftrue[::nmpts]
    finterp = lagran_interp(xvals,fvals,xinterp)
    error = np.ma.log10(np.abs(ftrue-finterp))
    return error 
    
# Error for Chebyshev interpolation
def cheb_spaced_error(nmpts,xinterp,ftrue,f):
    xvals = np.cos(np.pi*(2.*np.arange(nmpts+1)+1.)/(2.*nmpts+2.))
    fvals = f(xvals)
    finterp = lagran_interp(xvals,fvals,xinterp)
    error = np.ma.log10(np.abs(ftrue-finterp))
    return error 
    
# Common plotting routine for making 2x2 subplots 
def figure_maker(errors, npts, xinterp):
    nrows = 2
    ncols = 2
    fig, axs = plt.subplots(nrows, ncols, figsize=(9, 9), sharex=False)
    cnt = 0
    for jj in xrange(nrows):
        for kk in xrange(ncols):
            ax = axs[jj,kk]
            ax.plot(xinterp, errors[cnt], ls='-', color='k')
            ax.set_xlabel('$x$')
            ax.set_ylabel('$\log_{10}|Errors|$')
            ax.set_title('Interpolation over %d nodes' %npts[cnt])
            cnt += 1
    fig.tight_layout()
```

Using $n=10, ~20, ~60$ and $80$ equally spaced nodal points generates the error plots below.  As we see, the error is the greatest at the endpoints of the interval, though for $n=10$ and $n=20$, the overall error profile is relatively good with a maximum error of approximately $10^{-3}$ and $10^{-5}$ respectively.  However, once we set $n=60$, the error at the endpoints is in magnitude the same as that of the function, showing a relative error of almost 100 percent.  Using yet more points, as in $n=80$, exacerbates the error at the endpoints further.  Thus we can argue that the approximation method breaks down at $n=60$.  

```python
errors = []
npts = [10, 20, 60, 80]
for npt in npts:
    errors.append(equi_spaced_error(npt,xinterp,ftrue))
figure_maker(errors, npts, xinterp)
```

If we instead use Chebyshev spacing for our nodal points, repeating our experiments above, we get below the following error plots.  As can be seen, the overall error profile is significantly improved even for $n=10$ nodes.  Morever, by $n=60$, Chebyshev interpolation generates an approximation whose error is on the order of $10^{-15}$ throughout the entire interval $[-1,1]$, showing that machine precision has been reached.  This degree of accuracy persists when we choose $n=80$, in effect showing that adding more nodal points is superfluous.    

```python
errors = []
npts = [10, 20, 60, 80]
for npt in npts:
    errors.append(cheb_spaced_error(npt,xinterp,ftrue,f))
figure_maker(errors, npts, xinterp)
```

**Problem 2**: 10 pts - In class, we showed that if interpolated over three points, say $x_{j-1}$, $x_{j}$, $x_{j+1}$, with data $f_{j-1}$, $f_{j}$, and $f_{j+1}$, where our nodes are equally spaced so that 

$$
\delta x = x_{j+1} - x_{j} = x_{j} - x_{j-1}
$$

then using $f(x)\approx P_{2}(x)$, we have 

$$
f'(x_{j}) \approx \frac{f_{j+1}-f_{j-1}}{2\delta x}
$$

This is known as a _centered difference_ approximation to a derivative of a function.  

* (3 pts) Using the same interpolation information, find an approximation to $f''(x_{j})$.  
* (3 pts) Using $f(x) = \cos(x)$, determine the accuracy of your centered difference approximations with respect to the magnitude of $\delta x$ for both the first and second derivatives.  In other words, if we suppose that 
$$
f'(x_{j}) = \frac{f_{j+1}-f_{j-1}}{2\delta x} + C\delta x^{p},
$$
then find $p$.  Clearly explain your methodology.  
* (4 pts) Derive a centered difference approximation for $f'(x_{j})$ and $f''(x_{j})$ using the equally spaced nodes $x_{j-2}$, $x_{j-1}$, $x_{j}$, $x_{j+1}$, $x_{j+2}$, with corresponding node spacing $\delta x$, and corresponding data $f_{j-2}$, $f_{j-1}$, $f_{j}$, $f_{j+1}$, $f_{j+2}$.  Determine the order of accuracy of your approximation again using $f(x) = \cos(x)$.   


In order to measure the error in our approximation, we look at points $x_{j}\in[0, 2\pi]$ where $\delta x = 2\pi/N$ so that

$$
x_{j} = \frac{2\pi j}{N}, ~j=0, \cdots, N.
$$

We plot the maximum of the absolute value of the error in our approximation over our chosen interval for increasing choices of $N$.

```python
def fp(xvals,fvals):
    dx = xvals[1] - xvals[0]
    n = fvals.size
    fpv = (fvals[2:]-fvals[:n-2])/(2.*dx)
    err = np.max(np.abs(fpv+np.sin(xvals[1:n-1])))
    return err

def fpp(xvals,fvals):
    dx = xvals[1] - xvals[0]
    n = fvals.size
    fpv = (fvals[2:] - 2.*fvals[1:n-1] + fvals[:n-2])/dx**2.
    err = np.max(np.abs(fpv+np.cos(xvals[1:n-1])))
    return err

def err_plot(ffun):
    pows = np.arange(1,5)
    Npts = 10.**pows
    evec = np.zeros(Npts.size)
    cnt = 0
    for Npt in Npts:
        xvals = np.linspace(0.,2.*np.pi,int(Npt))
        fvals = np.cos(xvals)
        evec[cnt] = np.log10(ffun(xvals,fvals))
        cnt += 1
    
    slope = evec[1:]-evec[0:-1]
    print "Maximum of the slope is: %1.15f" % np.max(slope)
    print "Minimum of the slope is: %1.15f" % np.min(slope)
    print "Mean of the slope is: %1.15f" % np.mean(slope)
    plt.plot(np.log10(Npts), evec)
    plt.xlabel('$\log_{10}(N)$')
    plt.ylabel('$\log_{10}(Error)$')
    
```

```python
# Plot of the error in the approximation of the first derivative.  
err_plot(fp)
```

```python
# Plot of the error in the approximation of the second derivative.  
err_plot(fpp)
```

As we see from the plots, the order of accuracy of both of our approximations is $p=2$.

To derive approximations for derivatives using five data points, letting $u=x-x_{j}$, we see that the five basis Lagrange polynomials are

$$
L_{j-2}(u) = \frac{1}{24(\delta x)^{4}}\left(u+\delta x\right) u \left(u-\delta x \right)\left(u-2\delta x\right)
$$

$$
L_{j-1}(u) = -\frac{1}{6(\delta x)^{4}}\left(u+2\delta x\right) u \left(u-\delta x \right)\left(u-2\delta x\right)
$$

$$
L_{j}(u) = \frac{1}{4(\delta x)^{4}}\left(u+2\delta x\right) \left(u + \delta x\right)\left(u-\delta x \right)\left(u-2\delta x\right)
$$

$$
L_{j+1}(u) = -\frac{1}{6(\delta x)^{4}}\left(u+2\delta x\right) \left(u+\delta x \right)u\left(u-2\delta x\right)
$$

$$
L_{j+2}(u) = \frac{1}{24(\delta x)^{4}}\left(u+2\delta x\right) \left(u+\delta x \right)u \left(u-\delta x\right)
$$


Differentiating each of these and evaluating at $u=0$, or $x=x_{j}$, we find the approximations

$$
f'(x_{j}) \approx \frac{1}{12\delta x}\left(f_{j-2} - 8f_{j-1} + 8f_{j+1} - f_{j+2} \right)
$$

and

$$
f''(x_{j}) \approx \frac{1}{12(\delta x)^{2}}\left(-f_{j-2} + 16f_{j-1} - 30f_{j} + 16f_{j+1} - f_{j+2} \right)
$$

```python
def fp4(xvals,fvals):
    dx = xvals[1] - xvals[0]
    n = fvals.size
    fpv = (-fvals[4:]+8.*fvals[3:n-1]-8.*fvals[1:n-3]+fvals[:n-4])/(12.*dx)
    err = np.max(np.abs(fpv+np.sin(xvals[2:n-2])))
    return err

def fpp4(xvals,fvals):
    dx = xvals[1] - xvals[0]
    n = fvals.size
    fpv = (-fvals[4:]+16.*fvals[3:n-1]-30.*fvals[2:n-2]+16.*fvals[1:n-3]-fvals[:n-4])/(12.*dx**2.)
    err = np.max(np.abs(fpv+np.cos(xvals[2:n-2])))
    return err
```

```python
# Plot of the error in the approximation of the first derivative.  
err_plot(fp4)
```

```python
# Plot of the error in the approximation of the second derivative.  
err_plot(fpp4)
```

So as we see, by increasing the number of data points from three to five, and thus the order of the approximating polynomails from two to four, we improve the order of accuracy from $p=2$ to $p=4$, at least until various rounding effects take hold preventing further improvement.
