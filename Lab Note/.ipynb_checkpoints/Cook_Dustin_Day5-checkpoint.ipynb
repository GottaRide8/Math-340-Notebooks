```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def pc_func(x,tol):
    x2 = x**2.
    tot = 1.
    am = tot
    mm = 2.
    while np.abs(am) >= tol:
        am *= x2/((2.*mm+2)*(3.*mm+1.))
        tot += am
        mm += 1.
    return tot + 1.
```

```python
xvals = np.linspace(-10.,10.,int(1e3))

yvals = [pc_func(val,1e-8) for val in xvals]

plt.plot(xvals,yvals)
```
